```python
#Imports
import csv
import matplotlib.pyplot as plt
```

```python
#Function to read csv file of stellar interior data from starmodl
def read_csv(filename):
    
    #Reads csv file into lists
    with open(filename, 'r') as file:
        reader = csv.reader(file, delimiter=' ')
        data = list(reader)
        
    #Lists to store stellar interior data
    mass_list = []
    temp_list = []
    radius_list = []
    pressure_list = []
    luminosity_list = []
        
    #Loop to append stellar interior data into respective lists
    for row in data[16:]:
        mass_list.append(1-float(row[1]))
        temp_list.append(float(row[3]))
        radius_list.append(float(row[0]))
        pressure_list.append(float(row[4]))
        luminosity_list.append(float(row[2]))
            
    return mass_list, temp_list, radius_list, pressure_list, luminosity_list


#Runs read_csv to import data
mass_list, temp_list, radius_list, pressure_list, luminosity_list = read_csv('starmodl_0.5mass.csv')
mass_list2, temp_list2, radius_list2, pressure_list2, luminosity_list2 = read_csv('starmodl_1.0mass.csv')
mass_list3, temp_list3, radius_list3, pressure_list3, luminosity_list3 = read_csv('starmodl_2.0mass.csv')
mass_list4, temp_list4, radius_list4, pressure_list4, luminosity_list4 = read_csv('starmodl_4.0mass.csv')
mass_list5, temp_list5, radius_list5, pressure_list5, luminosity_list5 = read_csv('starmodl_8.0mass.csv')
mass_list6, temp_list6, radius_list6, pressure_list6, luminosity_list6 = read_csv('starmodl_10.0mass.csv')
```

```python
#Plotting pressure versus radius for each stellar model
plt.figure(figsize=(8,4), dpi=100)
plt.plot(radius_list, pressure_list, linestyle='solid', color='darkorange', label='0.5M$_{\odot}$')
plt.plot(radius_list2, pressure_list2, linestyle='dashed', color='black', label='1.0M$_{\odot}$')
plt.plot(radius_list3, pressure_list3, linestyle='solid', color='dodgerblue', label='2.0M$_{\odot}$')
plt.plot(radius_list4, pressure_list4, linestyle='solid', color='limegreen', label='4.0M$_{\odot}$')
plt.plot(radius_list5, pressure_list5, linestyle='solid', color='red', label='8.0M$_{\odot}$')
plt.plot(radius_list6, pressure_list6, linestyle='solid', color='mediumorchid', label='10.0M$_{\odot}$')
plt.grid(True, alpha=0.4, linestyle='dashed', color='darkgray')
plt.xlabel('Radius [m]')
plt.ylabel('Pressure [dynes / cm$^2$]')
plt.title('Stellar Interior Model - Pressure vs Radius')
plt.legend(prop={'size':12}, ncol=2, frameon=False)
plt.rcParams['legend.loc'] = 'upper right'
plt.show()
```

```python
#Plotting mass versus radius for each stellar model
plt.figure(figsize=(8,4), dpi=100)
plt.plot(radius_list, mass_list, linestyle='solid', color='darkorange', label='0.5M$_{\odot}$')
plt.plot(radius_list2, mass_list2, linestyle='dashed', color='black', label='1.0M$_{\odot}$')
plt.plot(radius_list3, mass_list3, linestyle='solid', color='dodgerblue', label='2.0M$_{\odot}$')
plt.plot(radius_list4, mass_list4, linestyle='solid', color='limegreen', label='4.0M$_{\odot}$')
plt.plot(radius_list5, mass_list5, linestyle='solid', color='red', label='8.0M$_{\odot}$')
plt.plot(radius_list6, mass_list6, linestyle='solid', color='mediumorchid', label='10.0M$_{\odot}$')
plt.grid(True, alpha=0.4, linestyle='dashed', color='darkgray')
plt.xlabel('Radius [m]')
plt.ylabel('Mass [M$_r$ / M$_{tot}$]')
plt.title('Stellar Interior Model - Mass vs Radius')
plt.legend(prop={'size':12}, ncol=2, frameon=False)
plt.rcParams['legend.loc'] = 'lower right'
plt.show()
```

```python
#Plotting luminosity versus radius for each stellar model
plt.figure(figsize=(8,4), dpi=100)
plt.plot(radius_list, luminosity_list, linestyle='solid', color='darkorange', label='0.5M$_{\odot}$')
plt.plot(radius_list2, luminosity_list2, linestyle='dashed', color='black', label='1.0M$_{\odot}$')
plt.plot(radius_list3, luminosity_list3, linestyle='solid', color='dodgerblue', label='2.0M$_{\odot}$')
plt.plot(radius_list4, luminosity_list4, linestyle='solid', color='limegreen', label='4.0M$_{\odot}$')
plt.plot(radius_list5, luminosity_list5, linestyle='solid', color='red', label='8.0M$_{\odot}$')
plt.plot(radius_list6, luminosity_list6, linestyle='solid', color='mediumorchid', label='10.0M$_{\odot}$')
plt.yscale('log')
plt.grid(True, alpha=0.4, linestyle='dashed', color='darkgray')
plt.xlabel('Radius [m]')
plt.ylabel('Luminosity [erg / s]')
plt.title('Stellar Interior Model - Luminosity vs Radius')
plt.legend(prop={'size':12}, ncol=2, frameon=False)
plt.rcParams['legend.loc'] = 'lower right'
plt.show()
```
